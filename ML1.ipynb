## 20. Постройте ядерную оценку плотности целевой переменной Price. Оцените визуально, напоминает ли полученный график нормальное распределение. Постройте двумерную ядерную оценку плотности для целевой переменной Price и признака HouseFloor, затем оцените визуально на каких этажах и по какой цене продаётся основная масса квартир.

```python
plt.figure(figsize=(20, 8))


sns.kdeplot(data3['Price'], fill=True, color='red', alpha=0.55)
plt.title('Kernel density estimate for the target variable Price')
plt.xlabel('Price')
plt.ylabel('Density')
plt.grid()
plt.show()


plt.figure(figsize=(20, 8))
sns.kdeplot(data=data3, x='Price', y='HouseFloor', fill=True, cmap='Reds', thresh=0, levels=35)
plt.title('Two-dimensional kernel density estimation for Price and HouseFloor')
plt.xlabel('Price')
plt.ylabel('HouseFloor')
plt.grid()
plt.show()
```
